```python
#import
from pymatgen.ext.matproj import MPRester

from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#pour télécharger le ﬁchier CIF du matériau
with MPRester("CfP3t0zy6KRi8xvk") as m:
    # Structure for material id
    struct = m.get_structure_by_material_id("mp-5967")
```

# Diffraction des rayons X du TiCoSb (mp-5967)

```python
xrdcal = XRDCalculator(1.54060)
print("longueur d’onde \u03BB:",xrdcal.wavelength,"Å")
```

## Affichage des pics

```python
print("Diffractogramme du NaRhO2 pour un rayonnement X incident CuKa de longueur d’onde λ=1.54060 Å")
xrdcal.show_plot(struct,two_theta_range=(0,90), annotate_peaks=True, ax=None, with_labels=True, fontsize=10)
xrdcal.show_plot(struct,two_theta_range=(20,45), annotate_peaks=True, ax=None, with_labels=True, fontsize=10)
```

```python
patt = xrdcal.get_pattern(struct)
hkls=patt.hkls
twotheta=patt.x
intensity=patt.y
```

 ## Analyse des 3 premiers pics

### Premier pic

```python
#Premier pic
print("Angle d'incidence : " + str(twotheta[0]) + "°")
print("Intensité : " + str(intensity[0]))
print("Indices hkl associés et sa multiplicité " + str(hkls[0]))
```
